```python
df = pd.read_csv("../../data/Credit.csv")
df.head(3)
```

```python
# Didn't exactly see how this was done in class. Hopefully this is good
p1 = sns.regplot(x='Limit', y='Age', data=df)
p1
```

```python
lin_reg = LinearRegression()

# Need to use df.drop, because it returns a df, where as iloc returns a col (np series)
good_df = df.drop('Age', axis=1)
good_df = good_df.drop('Unnamed: 0', axis=1)
good_df = good_df.drop('Income', axis=1)
good_df = good_df.drop('Rating', axis=1)
good_df = good_df.drop('Cards', axis=1)
good_df = good_df.drop('Education', axis=1)
good_df = good_df.drop('Gender', axis=1)
good_df = good_df.drop('Student', axis=1)
good_df = good_df.drop('Married', axis=1)
good_df = good_df.drop('Ethnicity', axis=1)
good_df = good_df.drop('Balance', axis=1)

lin_reg.fit(good_df, df.iloc[:, 1])  
```
